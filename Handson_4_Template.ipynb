## 6.1 Prime Number Checker
Create a program that checks whether a number is a prime number and displays its factors if it is not a prime number.

### Console:
```powershell
Prime Number Checker

Please enter an integer between 1 and 5000: 5
5 is a prime number.

Try again? (y/n): y

Please enter an integer between 1 and 5000: 6
6 is NOT a prime number.
It has 4 factors: 1 2 3 6

Try again? (y/n): y

Please enter an integer between 1 and 5000: 200
200 is NOT a prime number.
It has 12 factors: 1 2 4 5 8 10 20 25 40 50 100 200

Try again? (y/n): n

Bye!
```

### Specifications:
- A prime number is divisible by two factors (1 and itself). For example, 7 is a prime number because it is only divisible by 1 and 7.
- If the user enters an integer that's not between 1 and 5000, the program should display an error message.
- If the number is a prime number, the program should display a message.
If the number is not a prime number, the program should display a message. Then, it should display the number of factors for the number and a list of those factors.
- Store the factors for each number in a list.
- Use functions to organize the code for this program.


````python
### CODE HERE ###
### CODE HERE ###
AGAIN = -1

def validateInput(num:int):
    if (num <= 1 or num >= 5000): 
        print("Invalid integer. Please try again.")
        return AGAIN
    else:
        return num

def calculateFactors(num: int)->int:
    cnt = list()
    for i in range(1, num+1):
        if (num % i == 0): cnt.append(i)
    return cnt

def isPrime(num: int)->bool:
    factors = calculateFactors(num)
    if (len(factors) == 2): print("%d is a prime number."%num)
    else: 
        print("%d is NOT a prime number."%num)
        print("It has", len(factors) ,  "factors:", *factors, sep = " ")

def main():
    print("Prime Number Checker")

    while (True):
        print()
        num = AGAIN
        
        while (num == AGAIN):
            num = validateInput(int(input("Please enter an integer between 1 and 5000: ")))
        isPrime(num)
    
        print()
        if (input("Try again? (y/n): ") == 'n'): break

    print("\nBye!")
    
if __name__ == "__main__":
    main()
````

## 6.2 - Contact Manager
Create a program that a user can use to manage the primary email address and phone number for a contact.

### Console
```powershell
Contact Manager

COMMAND MENU
list - Display all contacts
view - View a contact
add - Add a contact
del - Delete a contact
exit - Exit program

Command: list
1. Guido van Rossum
2. Eric Idle

Command: view
Number: 2
Name: Eric Idle
Email: eric@ericidle.com
Phone: +44 20 7946 0958

Command: add
Name: Mike Murach
Email: mike@murach.com
Phone: 559-123-4567
Mike Murach was added.

Command: del
Number: 1
Guido van Rossum was deleted.

Command: list
1. Eric Idle
2. Mike Murach

Command: exit

```

### Specifications:
- Use a list of lists to store the data for the contacts. Provide starting data for two or more contacts.
- For the `view` and `del` commands, display an error message if the user enters and invalid contact number.
- When you exit the program, all changes that you made to the contact list are lost.




````python
### CODE HERE ###
def printMenu():
    print("""COMMAND MENU
list - Display all contacts
view - View a contact
add - Add a contact
del - Delete a contact
exit - Exit program""")
    
def listMenu(name: list):
    for num in range(len(name)):
        print("%d. %s"%(num + 1, name[num]))

def viewMenu(name: list, email: list, phone: list):
    num = int(input("Number: "))
    if num > len(name): 
        print("Out of range")
        return
    print("Name: %s"%name[num-1])
    print("Email: %s"%email[num-1])
    print("Phone: +%s"%phone[num-1])

def addMenu(name: list, email: list, phone: list):
    name.append(input("Name: "))
    email.append(input("Email: "))
    phone.append(input("Phone: "))

    print("%s was added."%name[-1])

def delMenu(name: list, email: list, phone: list):
    num = int(input("Number: "))
    if num > len(name): 
        print("Out of range")
        return
    
    print("%s was deleted."%name[num-1])
    
    name.pop(num-1)
    email.pop(num-1)
    phone.pop(num-1)

def Exit():
    return False

def main():
    loop = True
    name = ["Guido van Rossum", "Eric Idle"]
    email = ["guido@guidovanrossum.com","eric@ericidle.com"]
    phone = ["44 20 7946 0958","44 20 7946 0958"]

    print("Contact Manager\n")
    printMenu()
    print("")


    while loop:
        cmd = input("Command: ")
        if cmd.lower() == "list":listMenu(name)
        if cmd.lower() == "view":viewMenu(name, email, phone)
        if cmd.lower() == "add":addMenu(name, email, phone)
        if cmd.lower() == "del":delMenu(name, email, phone)
        if cmd.lower() == "exit": loop = Exit()
        print()

    print("Bye!")

if __name__ == "__main__":
    main()
````
